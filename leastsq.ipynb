```python
#Imports
import numpy as np
import scipy.linalg as la


#Plots
import matplotlib.pyplot as plt
import seaborn as sns

# Apply the default theme
sns.set_theme()
```

# (a)
We have the table of data in the next cell.

Using `numpy.linalg.lstsq()`, fit a straight line, a quadratic function, and a cubic function to these data. Plot the data and your fitted functions in a graph. It is **not** allowed to use numpy.polyfit, but you may have a look at the documentation to see some examples.

```python
#Data

t = np.array([3, 11, 29, 32, 47, 63, 73, 99])
w = np.array([74, 72, 52, 35, 37, 20, 19, 19])
```

```python
#Define the matrices
matrix_line = np.vstack([t, np.ones(len(t))]).T
matrix_quad = np.vstack([t**2,t, np.ones(len(t))]).T
matrix_cube = np.vstack([t**3, t**2,t, np.ones(len(t))]).T

#Define the coefficients
a_l, b_l = np.linalg.lstsq(matrix_line, w, rcond=None)[0]
a_q,b_q,c_q = np.linalg.lstsq(matrix_quad, w, rcond=None)[0]
a_c,b_c,c_c,d_c = np.linalg.lstsq(matrix_cube, w, rcond=None)[0]
```

```python
#Plot line

_ = plt.plot(t,w, 'o', label= 'Original', markersize=10)
_ = plt.plot(t, a_l*t + b_l, 'r', label='fitted')
_ = plt.legend()
_ = plt.xlabel('t')
_ = plt.ylabel('w')
plt.show()
```

```python
#Plot quadratic

quad = np.polyfit(t, w, 2)  #Sanity check for the solution
a_q2 = quad[0]
b_q2 = quad[1]
c_q2 = quad[2]

_ = plt.plot(t, a_q*t**2+ b_q*t + c_q, 'r', label='Quadratic')
_ = plt.plot(t, a_q2*t**2+ b_q2*t + c_q2, 'go', label='Scipy')
_ = plt.plot(t,w, 'o', label= 'Original', markersize=10)
_= plt.legend()
_ = plt.xlabel('t')
_ = plt.ylabel('w')
```

```python
#Plot cubic
cube = np.polyfit(t, w, 3)  #Sanity check for the solution
a_c2 = cube[0]
b_c2 = cube[1]
c_c2 = cube[2]
d_c2 = cube[3]

_ = plt.plot(t,w, 'o', label= 'Original', markersize=10)
_ = plt.plot(t, a_c*t**3 + b_c*t**2+ c_c*t + d_c, 'r', label='Cubic')
_ = plt.plot(t, a_c2*t**3 + b_c2*t**2+ c_c2*t + d_c2, 'go', label='Scipy')
_ = plt.xlabel('t')
_ = plt.ylabel('w')
_ = plt.legend()
```
